```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

imagenet_mean = np.array([0.485, 0.456, 0.406])
imagenet_std = np.array([0.229, 0.224, 0.225])
```

```python



title_dic = {'baseline': 'baseline', 
             'POS_A_mask_A': 'mask_A', 'POS_B_mask_B': 'mask_B', 'POS_AB_mask_AB': 'mask_AB', 
             'POS_AB_exchange_AB': 'exchange_AB', 
             'POS_A_random_A_same_task': 'random_A_same_task', 'POS_B_random_B_same_task': 'random_B_same_task',
             'POS_B_random_B_other_task_Flickr': 'B_random_B_other_task_Flickr', 'POS_B_random_B_other_task_random': 'B_random_B_other_task_random',
             'POS_A_random_A_other_task': 'random_A_other_task', 'POS_B_random_B_other_task': 'random_B_other_task',
             'POS_A_animeGAN_A': 'animeGAN_A', 'POS_B_animeGAN_B': 'animeGAN_B',
             'POS_A_random_A_other_task_coco_gt': 'random_A_other_task_coco_gt'}

exp_id = 'POS_B_animeGAN_B'

img = np.load(f'/hhd3/ld/data/Painter_root/lol_enhance/component_analysis/img2_prompt_{exp_id}.npy') * imagenet_std + imagenet_mean

tgt = np.load(f'/hhd3/ld/data/Painter_root/lol_enhance/component_analysis/tgt2_prompt_{exp_id}.npy') * imagenet_std + imagenet_mean
```

```python
test_img = np.concatenate((img, tgt), axis=1)
plt.axis('off')
plt.title(f'{title_dic[exp_id]}')
plt.imshow(test_img)
```
