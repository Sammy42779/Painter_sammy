```python
# plt.title(f'exp_{title_dic[exp_id]}_output')
# plt.figure(figsize=(4, 2))
# plt.imshow(plt.imread(output))
# plt.show()

import matplotlib.pyplot as plt
import numpy as np

# 假设我们有一个名为 image_list 的列表，其中包含6个图像数据
# 在这里，我们将创建6个随机图像作为示例
# image_list = [np.random.random((100, 100)) for _ in range(6)]

# 设置图形尺寸（可选）
plt.figure(figsize=(18, 3)) # 设置宽度为 18 英寸，高度为 3 英寸

# 按顺序横向绘制6个图像
for i, exp_id in enumerate(title_dic.keys(), start=1):
    # print(i)
    img = f'/hhd3/ld/data/ade20k/ade20k_seg_inference_painter_vit_large.pth_ADE_train_00009574_exp_{exp_id}/ADE_val_00000001.png'
    img_data = np.array(Image.open(img).convert('RGB'))
    plt.subplot(1, 6, i) # 创建 1x6 子图中的第 i 个子图
    plt.imshow(img_data, cmap='gray') # 绘制图像
    plt.title(f'{exp_id}') # 为子图添加标题（ID）
    plt.axis('off') # 关闭坐标轴

plt.title('ADE_val_00000001.png')

# 显示图像
plt.show()
```
