```python
loss1_list = []
kl_list = []

with open('/ssd1/ld/ICCV2023/Painter_sammy/Painter/eval/lol/loss.txt', 'r') as f:
    for line in f.readlines():
        line = line.strip('\n').split(',')
        loss1 = float(line[0].split('(')[-1])
        kl_loss = float(line[3].split(' ')[-1].split('(')[-1])
        print(loss1)
        loss1_list.append(loss1)
        kl_list.append(kl_loss)
        print(kl_loss)

plt.plot(loss1_list)
plt.show()
```
