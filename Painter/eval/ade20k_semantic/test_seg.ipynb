```python
import numpy as np
import matplotlib.pyplot as plt
# from constant_utils import *
```

```python

# attack_id = 'none'
# attack_id = 'attack_A'
attack_id = 'attack_AB'

epsilon = 8
i = 10


img_2 = np.load(f'/hhd3/ld/data/Painter_root/ade20k_semantic/change_A/attack_AB_8/img2_prompt.npy')
img_2_adv = np.load(f'/hhd3/ld/data/Painter_root/ade20k_semantic/change_A/attack_AB_8/img2_prompt_adv.npy') 
tgt_2 = np.load(f'/hhd3/ld/data/Painter_root/ade20k_semantic/change_A/attack_AB_8/tgt2_prompt.npy') 
tgt_2_adv = np.load(f'/hhd3/ld/data/Painter_root/ade20k_semantic/change_A/attack_AB_8/tgt2_prompt_adv.npy') 


test_img = np.concatenate((img_2, tgt_2, img_2_adv, tgt_2_adv), axis=1)
plt.axis('off')
plt.title(f'{attack_id}: perturbation={epsilon}/255')
plt.imshow(test_img)
```

```python
print(img_2.min(), img_2.max(), img_2_adv.min(), img_2_adv.max())
print(tgt_2.min(), tgt_2.max(), tgt_2_adv.min(), tgt_2_adv.max())
```

```python
print((img_2_adv - img_2).min(), (img_2_adv - img_2).max(),(img_2_adv - img_2).min() >= -epsilon/255, (img_2_adv - img_2).max() <= epsilon/255)
print((tgt_2_adv - tgt_2).min(), (tgt_2_adv - tgt_2).max(),(tgt_2_adv - tgt_2).min() >= -epsilon/255, (tgt_2_adv - tgt_2).max() <= epsilon/255)
```
